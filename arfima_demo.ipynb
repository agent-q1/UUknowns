```python
import numpy as np
from scipy.special import gamma
from copy import deepcopy
from arfima import arfima
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
```

# Brownian Motion

```python
T = 1000
af = arfima(Ap = [], d = 0, Aq = [1]*T)
x = af.gen(T = T)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
plt.plot(x)
plt.xlabel('time')
plt.title('time-series')
plt.grid()

plt.subplot(1, 2, 2)
num_traj = 10
acorr = pd.DataFrame(columns=['delay', 'ac'])
corr_len = 500
delay = np.empty((num_traj*corr_len))
ac = np.zeros((num_traj*corr_len))
for i in range(num_traj):
    x = af.gen(T = T)
    a_corr = np.correlate(x, x, mode="full")
    a_corr = a_corr[a_corr.size//2:]/ np.max(a_corr)
    delay_ = np.arange(0, len(x))
    delay[i*corr_len:(i+1)*corr_len] = delay_[:corr_len]
    ac[i*corr_len:(i+1)*corr_len] = a_corr[:corr_len]
    
sns.lineplot(x="delay", y="auto-correlation", data=pd.DataFrame({'delay':delay, 'auto-correlation':ac}))
plt.title('Normalized auto-correlation')
plt.xlabel('delay')
plt.show()
```
